# Exploratory Data Analysis

### Segment

```python
# plotting a bar chart for Segment and it's count

ax = sns.countplot(x = 'Segment',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for Segment vs total amount

sales_seg = df.groupby(['Segment'], as_index=False)['Profit'].sum().sort_values(by='Profit', ascending=False)

sns.barplot(x = 'Segment',y= 'Profit' ,data = sales_seg)
```

*From above graphs we can see that most of the buyers are Consumer and even the profit from Consumer are greater than Corporate or Home Office*

### Region

```python
ax = sns.countplot(data = df, x = 'Region', hue = 'Segment')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Profit vs Region
sales_reg = df.groupby(['Region'], as_index=False)['Profit'].sum().sort_values(by='Profit', ascending=False)

sns.barplot(x = 'Region',y= 'Profit' ,data = sales_reg)
```

*From above graphs we can see that most of the buyers are from West Region*

### State

```python
# total Quantity of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Quantity'].sum().sort_values(by='Quantity', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Quantity')
```

```python
# total Profit from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Profit'].sum().sort_values(by='Profit', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Profit')
```

*From above graphs we can see that most of the orders & total sales are from California, New Work, Texas respectively but profit are from California, New Work, Washington respectively*


*So from above graphs we can oberserve that even though the sales are more in texas but its not one of top 10 profititable state*


*It is clear that not just sales matters but also budget of the buyers and price of the product*

### Category

```python
ax = sns.countplot(data = df, x = 'Category')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Category', 'Segment'], as_index=False)['Profit'].sum().sort_values(by='Profit', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Category',y= 'Profit', hue='Segment')
```

*From above graphs we can see that most of the consumed category is office supplies but profits from Technology is higher And nevertheless Consumer give higher profit*

### Shipping Mode

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Ship Mode')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Ship Mode'], as_index=False)['Profit'].sum().sort_values(by='Profit', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Ship Mode',y= 'Profit')
```

*From above graphs we can see that most of the product are shipped by Standard Class*

### Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Sub-Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Sub-Category'], as_index=False)['Profit'].sum().sort_values(by='Profit', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Sub-Category',y= 'Profit')
```

*From above graphs we can see that most of the sold products are from Binders, Papers and Furnishers category but Copiers has most profits*

```python
sales_state = df.groupby(['Product ID'], as_index=False)['Quantity'].sum().sort_values(by='Quantity', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product ID',y= 'Quantity')
```

```python
# top 10 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product ID')['Quantity'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
